```python
plt.scatter(x[y_predict==0,0],x[y_predict==0,1],s=100,c='coral' ,label='cluster 1')
plt.scatter(x[y_predict==1,0],x[y_predict==1,1],s=100,c='darksalmon' ,label='cluster 2')
plt.scatter(x[y_predict==2,0],x[y_predict==2,1],s=100,c='tomato' ,label='cluster 3')
plt.scatter(x[y_predict==3,0],x[y_predict==3,1],s=100,c='saddlebrown' ,label='cluster 4')
plt.scatter(x[y_predict==4,0],x[y_predict==4,1],s=100,c='peru' ,label='cluster 5')
plt.title('clusters of customers')
plt.xlabel('Anuual Income')
plt.ylabel('spending Score')
plt.legend()
plt.show()
```

```python
inertia=[]
k_range=range(2,11)
for i in k_range:
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)
    
fig,ax=plt.subplots(figsize=(10,8))
sns.lineplot(x=list(k_range),y=inertia)
plt.title('The Elbow Method')
plt.xlabel('No Of Clusters')
plt.ylabel('Inertia')
plt.show()
```
